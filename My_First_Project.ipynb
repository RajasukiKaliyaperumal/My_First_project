```python
import pandas as pd


data = {
    'Movie_Name': ['Movie_A', 'Movie_B', 'Movie_C', 'Movie_D', 'Movie_E'],
    'Budget': [10, 20, 30, 40, 50],
    'Collection': [30, 50, 70, 90, 110]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.describe())
```

```python
import matplotlib.pyplot as plt

df.plot(x='Budget', y='Collection', kind='line', marker='o')
plt.xlim(xmin=0, xmax=60)
plt.ylim(ymin=0, ymax=150)
plt.grid(True)
plt.show()
```

```python
import numpy as np
slope, intercept = np.polyfit(df['Budget'], df['Collection'], 1)
new_budget = 70
predicted_collection = ( slope * new_budget) + intercept
print(predicted_collection)
```
